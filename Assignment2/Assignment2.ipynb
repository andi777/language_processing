```python
import codecs
from bs4 import BeautifulSoup
import re
from nltk.tokenize import regexp_tokenize
from collections import Counter
```

# 4.	Create Python files:



```python
F = codecs.open("itwiki-latest-pages-articles.xml/smallCorpus.xml", encoding="UTF-8") 
contents = F.read() # contents if of type String
print (contents) 
```

# Clean the data:

## strip away all HTML tags.

```python
# nltk.clean_html() does not work; have to use BeautifulSoup

soup = BeautifulSoup(contents)
text = soup.get_text()
print(text)
```

### translate the most important html entities like this:
### •	&gt  ––>  >
### •	&lt  ––>  <
### •	&quot  ––>  "
### •	&amp  ––>  &


```python
regex = r">"
text = re.sub(regex, r"&gt", text)

regex = r"<"
text = re.sub(regex, r"&lt", text)

regex = r'"'
text = re.sub(regex, r"&quot", text)

regex = r"&"
text = re.sub(regex, r"&amp", text)
print(text)
```

### Remove the wiki formatting, which is described at the wikipedia cheatsheet and the wiki markup help page. Here are the most important things you have to do:
##### •	headings; remove all ==, ===, ====, etc. that occur both at the beginning and at the end of a line
##### •	list bullets, indendation; remove all *, **, ***, #, ##, ###, :, ::, :::, etc. at the beginning of a line
##### •	bold and italic; remove all apostrophe sequences (i.e., two or more ' after another)
##### •	remove everything surrounded by double braces: "{{...}}"
##### •	transform internal links "[[...]]" in the following way:
###### o	remove all "[[xx:...]]" where xx is a Wikipedia language code (either see the list of wikipedias, or use any 2 or 3 ascii characters a–z)
###### o	replace "[[...|" with "[["
###### o	replace "[[...:" with "[["
###### o	replace "[[text]]" with "text"


```python
regex = r"={1,4}"
text = re.sub(regex, r" ", text)

regex = r"#{1,3}"
text = re.sub(regex, " ", text)

regex = r":{1,3}"
text = re.sub(regex, " ", text)

regex = r"'{2,3}"
text = re.sub(regex, " ", text)

regex = r"(\{)(\{)(([\w\s](.)+))(\})(\})"
text = re.sub(regex, " ", text)

regex = r"\[([A-Za-z0-9_|.]+)\]"
text = re.sub(regex, " ", text)

regex = r"([[])"
text = re.sub(regex, "", text)

regex = r"([]])"
text = re.sub(regex, "", text)
```

```python
print(text)
```

```python
tokens = regexp_tokenize(text, pattern='\w+')
print (tokens)
```

### How big is the corpus (# of words)?

```python
wordcount = len(re.findall(r'\w+', text))
print (wordcount)
```

### What is the average word length? 

```python
w = [len(word) for line in tokens for word in line.rstrip().split(" ")]
w_avg = float(sum(w))/float(len(w))
print ('What is the average word length? : ', w_avg)
```

### What are the 10 most frequent words? How many percent of the corpus consists of these words?

```python
listOfWordsAndCounts = []

for word in tokens:
    wordList = re.findall(word, text)
    wordLenght = len(wordList) #find length of wordList
    listOfWordsAndCounts.append((word, wordLenght)) #add word and wordLenght to a dictionary

listOfWordsAndCounts = sorted(listOfWordsAndCounts,  key=lambda x: x[1])
print ('What are the 10 most frequent words? : ',listOfWordsAndCounts[:-10:-1])
```
